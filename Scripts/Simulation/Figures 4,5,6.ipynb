```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

obs = [100, 200, 400, 800, 1600, 3200]
```

```python
GRF_x = {n: pd.read_csv(f'Estimates {n}x25\\step2_GRF_x.csv') for n in obs}
GRF_d = {n: pd.read_csv(f'Estimates {n}x25\\step2_GRF_d.csv') for n in obs}
MLP   = {n: pd.read_csv(f'Estimates {n}x25\\step2_MLP.csv')   for n in obs}
```

```python
gridRFx = np.array([(x,y) for x in range(2) for y in np.linspace(0.1, 5, 50)])
gridRFd = np.array([(x,y) for x in np.linspace(0.1, 5, 50) for y in np.linspace(0.1, 5, 50)])
gridNN  = np.array([(x,y,z) for x in range(2) for y in np.arange(1, 6, 2) for z in np.arange(0.1, 5, 0.1)])
indices = [(i,j) for i in range(3) for j in range(2)]
```

```python
fig, ax = plt.subplots(nrows = 3, ncols = 2, figsize = (15, 20), sharey = True)
for i, n in enumerate(obs):
    for x_ in range(2):
        ax[indices[i]].plot(gridRFx[(gridRFx[:,0] == x_),1],
                            np.mean(GRF_x[n], axis = 0)[(gridRFx[:,0] == x_)],
                            label = '$d = {0}$'.format(x_))
        ax[indices[i]].fill_between(gridRFx[(gridRFx[:,0] == x_),1],
                                    np.percentile(GRF_x[n], 2.5, axis = 0)[(gridRFx[:,0] == x_)].reshape(-1),
                                    np.percentile(GRF_x[n], 97.5, axis = 0)[(gridRFx[:,0] == x_)].reshape(-1),
                                    alpha = 0.1)
        ax[indices[i]].axhline(0, color = 'grey', linewidth = 0.5)
        ax[indices[i]].set_title(f'Sample size = {n * 25:,}\n({n:,} markets)')
        ax[indices[i]].set(xlabel = '$x^2$', ylabel = '$\hat{CATE}$')
        ax[indices[i]].legend(loc = 'lower right')
        plt.ylim(top = 1.5, bottom = -1)
        [ax[indices[i]].axhline((np.linalg.inv(np.column_stack((np.ones(10**3), np.linspace(0.1, 5, 10**3))).T @ np.column_stack(((np.ones(10**3), np.linspace(0.1, 5, 10**3))))) @ (np.column_stack((np.ones(10**3), np.linspace(0.1, 5, 10**3))).T @ (np.log(np.linspace(0.1, 5, 10**3)).reshape(-1,1) * L)))[1,:],
                                color = 'black', ls = 'dashed', linewidth = 3)
         for L in [1,2]]
```

```python
fig, ax = plt.subplots(nrows = 3, ncols = 2, figsize = (15, 20), sharey = True)
for i, n in enumerate(obs):
    for x_ in [2,4]:
        ax[indices[i]].plot(gridRFd[gridRFd[:,0] == x_,1],
                            np.mean(GRF_d[n], axis = 0)[gridRFd[:,0] == x_],
                            label = '$x^2 = {0}$'.format(x_))
        ax[indices[i]].fill_between(gridRFd[gridRFd[:,0] == x_,1],
                                    np.percentile(GRF_d[n], 2.5, axis = 0)[(gridRFd[:,0] == x_)],
                                    np.percentile(GRF_d[n], 97.5, axis = 0)[(gridRFd[:,0] == x_)],
                                    alpha = 0.1)
        ax[indices[i]].axhline(0, color = 'grey', linewidth = 0.5)
        ax[indices[i]].set_title(f'Sample size = {n * 25:,}\n({n:,} markets)')
        ax[indices[i]].set(xlabel = '$x^1$', ylabel = '$\hat{CATE}$')
        ax[indices[i]].legend(loc = 'lower right')
        ax[indices[i]].set_ylim([-2, 2])
        [ax[indices[i]].plot(np.linspace(0.1, 5, 50),
                             np.log(np.linspace(0.1, 5, 50)) + np.log(L) - 1,
                             color = 'black', ls = 'dashed', linewidth = 2)
         for L in [2,4]]
```

```python
fig, ax = plt.subplots(nrows = 3, ncols = 2, figsize = (15, 20), sharey = True)
for i, n in enumerate(obs):
    for x_, y_ in [(x_,y_) for x_ in range(2) for y_ in np.arange(1, 6, 2)]:
        ax[indices[i]].plot(gridNN[(gridNN[:,0] == x_) & (gridNN[:,1] == y_),2] + 0.05,
                            np.mean(np.array(MLP[n]), axis = 0)[(gridNN[:,0] == x_) & (gridNN[:,1] == y_)],
                            label = f'$d = {x_}, x^2 = {y_}$')
        ax[indices[i]].fill_between(gridNN[(gridNN[:,0] == x_) & (gridNN[:,1] == y_),2] + 0.05,
                                    np.percentile(np.array(MLP[n]).reshape(-1, gridNN.shape[0]), 2.5, axis = 0)[(gridNN[:,0] == x_) & (gridNN[:,1] == y_)],
                                    np.percentile(np.array(MLP[n]).reshape(-1, gridNN.shape[0]), 97.5, axis = 0)[(gridNN[:,0] == x_) & (gridNN[:,1] == y_)],
                                    alpha = 0.1)
        ax[indices[i]].set_title(f'Sample size = {n * 25:,}\n({n:,} markets)')
        ax[indices[i]].set(xlabel = '$x^1$', ylabel = 'Marginal Effect')
        ax[indices[i]].set_ylim(top = 10)
        ax[indices[i]].legend(loc = 'upper right')
        ax[indices[i]].axhline(0, color = 'grey', linewidth = 0.5)
        [ax[indices[i]].plot(gridNN[(gridNN[:,0] == x_) & (gridNN[:,1] == y_),2] + 0.05,
                             L / (gridNN[(gridNN[:,0] == x_) & (gridNN[:,1] == y_),2] + 0.05),
                             color = 'black', ls = 'dashed', linewidth = 2)
         for L in [1,2]]
```
